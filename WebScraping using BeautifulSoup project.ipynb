# 1> Write a python program to display all the header tags from 
www.wikipedia.org

```python
url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
url
```

# 2>Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url=requests.get('https://www.imdb.com/chart/top/')
url
```

```python
soup=BeautifulSoup(url.content)
```

```python
movie_name = []

for i in soup.find_all('td',class_="titleColumn"):
    movie_name.append(i.get_text().replace('\n', ""))

movie_name
```

```python
yearofrel = []

for i in soup.find_all('span',class_="secondaryInfo"): 
    yearofrel.append(i.text)

yearofrel
```

```python
ratings = []

for i in soup.find_all('td',class_="ratingColumn imdbRating"): 
    ratings.append(i.text)

ratings
```

```python
data=pd.DataFrame()
data["Movie Name"]=movie_name
data["Year Of Release"]=yearofrel
data["Ratings"]=ratings
data.to_csv("IMDB’s Top rated movies",index=False)
```

```python
pd.read_csv("IMDB’s Top rated movies")
```

# 7> Write a python program to scrape mentioned news details from 
https://www.cnbc.com/world/?region=world

```python
url=requests.get('https://www.cnbc.com/world/?region=world')
url
```

```python
soup=BeautifulSoup(url.content)
```

```python
Headlines = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    Headlines.append(i.get_text().replace('\n', ""))
    
Headlines
```

```python
time = []

for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.get_text().replace('\n', ""))
    
time
```

```python
NewsLink = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    NewsLink.append(i.get("href").replace("\n", " "))
    
NewsLink
```

```python
ratings = []

for i in soup.find_all('div',class_="restnt-rating rating-4"): 
    ratings.append(i.text)

ratings
```

```python
images = []

for i in soup.find_all('img',class_="no-img"): 
    images.append(i.get('data-src'))

images
```
